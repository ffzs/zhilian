```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("job_0.2.csv")

df = data.iloc[:,[0,2,3,4,5,7,8,13]]

df.head()
```

```python
df = df.replace({'score':"-"},50)
df.dtypes
```

```python
df["score"] = df["score"].astype("int")
df.dtypes
```

```python
df = df.rename(columns={"参考平均工资":"salary","工作名称":"job_name","工作标签数":"tags","工作详情":"job_info"})
df.tail(3)
```

```python
df.to_csv("job_0.3.csv",encoding="utf-8")
```

```python
def name_score(x):
    name_ban = ["linux","php","Linux","PHP","JAVA","java","DBA","运维","web","WEB","实习生","C","C++","培训","讲师","R"]
    grade = 0
    for name in name_ban:
        if name in x:
            x = ""
    name_up = ["python","Python","爬虫","数据分析","机器学习","生物"]
    for name in name_up:
        if name in x:
            grade +=4
    name_down = ["自然语言","金融","安全"]
    for name in name_down:
        if name in x:
            grade -=2
    if x =="" :
        return np.nan
    else :
        return grade
```

```python
df["name_grade"] = df["job_name"].apply(name_score).dropna()
```

```python
df.head()
```

```python
def info_score(x):
    grade = 0.0
    name_up = ["python","Python","爬虫","数据分析","机器学习","生物","挖掘","量化"]
    for name in name_up:
        if name in x:
            grade +=1.0
    name_down = ["博士","硕士","自然语言","金融","安全","linux","php","Linux","PHP","JAVA","java","DBA","运维","web","WEB","实习生","C","C++","培训","讲师","R"]
    for name in name_down:
        if name in x:
            grade -=1.0
    if x =="" :
        return np.nan
    else :
        return grade
```

```python
df["info_grade"] = df["job_info"].apply(info_score).dropna()
df.head()
```

```python
df["final_score"] = df["score"]/10.0 + df["name_grade"] + df["info_grade"]+df["salary"]/2000.0
df.to_csv("job_0.5.csv",encoding="utf-8")
df.head()
```

```python
df_count = df.groupby("company").count().iloc[:,[0]].rename(columns={"salary":"count"})
df_count.tail(3)
```

```python
df_mean = df.groupby("company").mean().dropna().iloc[:,2:4]
df_mean.tail(3)
```

```python
df_max = df.groupby("company").max().dropna().iloc[:,[-1]]
df_max.head()
```

```python
df_final = pd.merge(df_count,df_max,right_index=True,left_index=True)
df_final = pd.merge(df_final,df_mean,right_index=True,left_index=True)
df_final.head()
```

```python
df_final.to_csv("job_1.0.csv",encoding="utf-8")
```
